```python
#Install and import requests library
!pip3 install requests
import requests
```

```python
#Extract content from web
url = 'https://portal.inmet.gov.br/dadoshistoricos'
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.102 Safari/537.36 Edg/104.0.1293.63'
}
response = requests.get(url, headers=headers)
html_bytes = response.content
```

```python
#Check if importation went well
response
```

```python
#Install and import BS library
!pip3 install bs4
from bs4 import BeautifulSoup as bs
```

```python
#Define soup variable 
soup = bs(html_bytes)
```

```python
#Check if soup is right
print(soup.prettify())
```

```python
#Create a list with all urls of csv files to download later.
tag_a = soup.find_all('article', attrs = {'class':'post-preview'})
```

```python
#Check if all of then are in the list
len(tag_a)
```

```python
#Clean the list, to content just the urls
urls = []
for elemento in tag_a:
    urls.append(elemento.find('a').attrs['href'])
print(urls)
len(urls)
```

```python
#Download files from urls
for url in urls:
    file_name = url[-8:]
    response = requests.get(url)
    open(file_name, "wb").write(response.content)
```

```python
# Importing required python libraries
import os, glob
import pandas as pd
# Changing the current working directory
os.chdir(r"C:\Users\yasmi\Documents\Ironhack - DA 2022\Módulo 2 - SQL\Project\Parte 1\all_files")

# Getting the FileNames of all .csv files in the current dir.
filenames = [i for i in glob.glob(f"*.csv")] 

# Loading all the csv files to create a list of data frames
df = [pd.read_csv(file, na_values = -9999, skiprows = 8, sep = ';', encoding = 'latin-1', engine = 'python' ) 
      for file in filenames]
```

```python
#Count total rows to check if the final table is correct
rows = 0
for x,y in shape_dfs:
        rows += x
        
print(rows)
```

```python
#Concat tables
tb_pluv = pd.concat(df)
```

```python
#Drop columns
tb_pluv = tb_pluv[['DATA (YYYY-MM-DD)','HORA (UTC)','PRECIPITAÇÃO TOTAL, HORÁRIO (mm)','TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)']]
tb_pluv
```

```python
#Rename columns
tb_pluv.columns = ['Data','Hora','Precipitação total (mm)', 'Temperatura do ar (°C)']
tb_pluv
```

```python
#Some pluviometric values starts with ",". Create a mask to drop these commas later.
mask_startscomma = [str(x).startswith(',') for x in tb_pluv['Precipitação total (mm)']]
```

```python
#Create a list with the pluviometric data corrected, to replace the column "Precipitação total (mm)" later
precipitação = []
for elemento in tb_pluv['Precipitação total (mm)']:
    if str(elemento).startswith(','):
        precipitação.append(elemento.replace(',','', 1))
    else:
        precipitação.append(elemento)
```

```python
#Check if the list has all the lines of the tb_pluv
len(precipitação)
```

```python
#Replace column "Precipitação total (mm)" to the corrected values from the list "precipitação"
tb_pluv['Precipitação total (mm)'] = precipitação
```
